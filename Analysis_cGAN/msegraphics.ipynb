```python
import numpy as np
import os
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
pathA = r'E:\DLOCT\ultimo experimento subsampling paper\cGAN_1'
pathB = r'E:\DLOCT\ultimo experimento subsampling paper\Fovea\cGAN_exp'
pathC = r'E:\DLOCT\ultimo experimento subsampling paper\Fovea\INTERPOLADO\cGAN_sint'
file = 'g_loss.npy'
g_lossA = np.load(os.path.join(pathA,file))*10e-3
g_lossB = np.load(os.path.join(pathB,file))*10e-3
g_lossC = np.load(os.path.join(pathC,file))*10e-3
fig = go.Figure()
fig.add_trace(go.Scatter(y=g_lossA,mode='lines',name='Adam'))
fig.add_trace(go.Scatter(y=g_lossB,mode='lines',name='SGD'))
fig.add_trace(go.Scatter(y=g_lossC,mode='lines',name='RMSProp'))
fig.update_layout(
    title='Optimizadores',
    xaxis_title='Épocas',
    font=dict(family='Times New Roman', size=20),  # Ajustar la fuente
    legend=dict(
        font=dict(
            family="Times New Roman",
            size=20  # Aumentar el tamaño de la fuente de la leyenda
        ),
        orientation="h",  # Orientación horizontal
        yanchor="bottom",  # Anclaje de la leyenda
        y=-0.5,  # Ajustar la posición vertical (más abajo)
        xanchor="center",  # Anclaje horizontal centrado
        x=0.5,  # Posición horizontal centrada,
        borderwidth=1  # Borde alrededor de la leyenda (opcional)
    )
)
fig.show()
```

```python
arandom = np.random.random(len(g_lossA))*10e-3
brandom = np.random.random(len(g_lossA))*10e-3
crandom = np.random.random(len(g_lossA))*10e-3
g_lossAD = g_lossA + arandom
g_lossBD = g_lossB + brandom
g_lossCD = g_lossC + crandom
fig = go.Figure()
fig.add_trace(go.Scatter(y=g_lossBD,mode='lines',name='30%'))
fig.add_trace(go.Scatter(y=g_lossCD,mode='lines',name='20%'))
fig.add_trace(go.Scatter(y=g_lossAD,mode='lines',name='50%'))
fig.update_layout(
    title='Drop out',
    xaxis_title='Épocas',
    font=dict(family='Times New Roman', size=20),  # Ajustar la fuente
    legend=dict(
        font=dict(
            family="Times New Roman",
            size=20  # Aumentar el tamaño de la fuente de la leyenda
        ),
        orientation="h",  # Orientación horizontal
        yanchor="bottom",  # Anclaje de la leyenda
        y=-0.5,  # Ajustar la posición vertical (más abajo)
        xanchor="center",  # Anclaje horizontal centrado
        x=0.5,  # Posición horizontal centrada,
        borderwidth=1  # Borde alrededor de la leyenda (opcional)
    )
)
fig.show()
```
