```python
#%%
import matplotlib.pyplot as plt
import numpy as np
import os
path = r'C:\Users\USER\OneDrive - Universidad EAFIT\Eafit\Trabajo de grado\Bibliografía complex conjugate mirror terms\imagenes_tesis'


# Parámetros del haz
radius = 1  # Radio del haz
distance_overlap = 0.5  # Distancia para el solapamiento (menor que el diámetro para que se solapen)
distance_no_overlap = 1  # Distancia para los haces separados (mayor que el diámetro)

# Crear la figura y ejes
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Función para dibujar un haz con gradiente
def draw_beam(ax, center, radius, alpha=0.3, color='red'):
    circle = plt.Circle(center, radius, color=color, alpha=alpha)
    ax.add_patch(circle)

# Gráfico para haces solapados
ax1 = axes[0]
ax1.set_aspect('equal')
ax1.set_xlim(-3, 3)
ax1.set_ylim(-1.5, 1.5)
ax1.axis('off')

# Dibujar dos haces solapados
draw_beam(ax1, (-distance_overlap, 0), radius, alpha=0.5)
draw_beam(ax1, (distance_overlap, 0), radius, alpha=0.5)
# ax1.set_title('Haces Solapados')

# Gráfico para haces no solapados
ax2 = axes[1]
ax2.set_aspect('equal')
ax2.set_xlim(-3, 3)
ax2.set_ylim(-1.5, 1.5)
ax2.axis('off')

# Dibujar dos haces separados
draw_beam(ax2, (-distance_no_overlap, 0), radius, alpha=0.5)
draw_beam(ax2, (distance_no_overlap, 0), radius, alpha=0.5)
# ax2.set_title('Haces No Solapados')

plt.tight_layout()
plt.show()
fig.savefig(os.path.join(path,'haces_individuales.svg'), format='svg')
```

```python
# Parámetros del haz
radius = 1  # Radio del haz
distance_overlap_2d = 1  # Distancia para el solapamiento (menor que el diámetro para que se solapen)
distance_no_overlap_x = 2.01  # Distancia para los haces no solapados en el eje X
distance_no_overlap_y = 1  # Distancia para los haces solapados en el eje Y
grid_size = 5  # Tamaño de la cuadrícula de haces para solapados
grid_size_x = 3  # Tamaño de la cuadrícula de haces para no solapados (X)
grid_size_y = 5  # Tamaño de la cuadrícula de haces para no solapados (Y)

# Crear la figura y ejes
fig, axes = plt.subplots(1, 2, figsize=(10, 5))


def draw_beam_grid(ax, distance, radius, grid_size, alpha=0.3, color='red'):
    for i in range(grid_size):
        for j in range(grid_size):
            center = (i * distance, j * distance)
            circle = plt.Circle(center, radius, color=color, alpha=alpha)
            ax.add_patch(circle)


def draw_custom_grid(ax, radius, distance_x, distance_y, grid_size_x, grid_size_y, alpha=0.4, color='red'):
    for i in range(grid_size_x):
        for j in range(grid_size_y):
            center = (i * distance_x, j * distance_y)
            circle = plt.Circle(center, radius, color=color, alpha=alpha)
            ax.add_patch(circle)


ax1 = axes[0]
ax1.set_aspect('equal')
ax1.set_xlim(-1, grid_size * distance_overlap_2d)
ax1.set_ylim(-1, grid_size * distance_overlap_2d)
ax1.axis('off')

# Dibujar cuadrícula solapada
draw_beam_grid(ax1, distance_overlap_2d, radius, grid_size, alpha=0.4)
# ax1.set_title('Cuadrícula de Haces Solapados')


ax2 = axes[1]
ax2.set_aspect('equal')
ax2.set_xlim(-1, grid_size_x * distance_no_overlap_x)
ax2.set_ylim(-1, grid_size_y * distance_no_overlap_y)
ax2.axis('off')

# Dibujar cuadrícula con solapamiento parcial en un solo eje
draw_custom_grid(ax2, radius, distance_no_overlap_x, distance_no_overlap_y, grid_size_x, grid_size_y)
# ax2.set_title('Cuadrícula de Haces con Solapamiento Parcial')

plt.tight_layout()
plt.show()
fig.savefig(os.path.join(path,'haces_rejillas.svg'), format='svg')

```
